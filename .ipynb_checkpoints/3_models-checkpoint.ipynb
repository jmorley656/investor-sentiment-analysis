### Logistic Regression on a TFIDF Vectorizer

### Random Forest model (CV)

```python
# Setting the count vectorizer
cvec = CountVectorizer()

# Fitting the vectorizer to the training data
cvec.fit(X)
```

```python
Xcv_df = pd.DataFrame(cvec.transform(X).toarray(), 
                      columns=cvec.get_feature_names_out())
```

```python
Xcv_train, Xcv_test, ycv_train, ycv_test = train_test_split(
    Xcv_df, y, test_size=0.2, random_state=1)
```

```python
rf_cv = RandomForestClassifier(n_estimators=100,
                                  random_state=1)
rf_cv.fit(Xcv_train, ycv_train)
```

```python
print('CV Random Forest Model')
print('****************************')
print('Baseline Accuracy:', round(bow.op_bin.value_counts(normalize=True)[1], 4))
print('Training data score:', round(rf_cv.score(Xcv_train, ycv_train), 4))
print('Mean CV score:', round(cross_val_score(rf_cv, Xcv_train, ycv_train, cv=5).mean(), 4))
print('Test data score:', round(rf_cv.score(Xcv_test, ycv_test), 4))
```

### Random Forest Model (TFIDF)

```python
# Setting the vectorizer
tvec = TfidfVectorizer()

# Fitting the vectorizer to the training data
tvec.fit(X)
```

```python
Xtv_df = pd.DataFrame(tvec.transform(X).toarray(), 
                      columns=tvec.get_feature_names_out())
```

```python
Xtv_train, Xtv_test, ytv_train, ytv_test = train_test_split(
    Xtv_df, y, test_size=0.2, random_state=1)
```

```python
rf_tv = RandomForestClassifier(n_estimators=100,
                                  random_state=1)
rf_tv.fit(Xtv_train, ytv_train)
```

```python
print('TFIDF Random Forest Model')
print('****************************')
print('Baseline Accuracy:', round(bow.op_bin.value_counts(normalize=True)[1], 4))
print('Training data score:', round(rf_tv.score(Xtv_train, ytv_train), 4))
print('Mean CV score:', round(cross_val_score(rf_tv, Xtv_train, ytv_train, cv=5).mean(), 4))
print('Test data score:', round(rf_tv.score(Xtv_test, ytv_test), 4))
```
